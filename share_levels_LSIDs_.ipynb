```python
# List of base file names
file_base_names = [
    "Brighton_and_Hove_East_Sussex_West_Sussex_LSIP_Main_Tables",
    "Buckinghamshire_LSIP_Main_Tables",
    "Cambridgeshire_and_Peterborough_LSIP_Main_Tables",
    "Cheshire_and_Warrington_LSIP_Main_Tables",
    "Cornwall_and_the_Isles_of_Scilly_LSIP_Main_Tables",
    "Cumbria_LSIP_Main_Tables",
    "Derbyshire_and_Nottinghamshire_LSIP_Main_Tables",
    "Dorset_LSIP_Main_Tables",
    "Essex_Southend-on-Sea_and_Thurrock_LSIP_Main_Tables",
    "Gloucestershire_LSIP_Main_Tables",
    "Greater_Lincolnshire_LSIP_Main_Tables",
    "Greater_London_LSIP_Main_Tables",
    "Greater_Manchester_LSIP_Main_Tables",
    "Heart_of_the_South-West_LSIP_Main_Tables",
    "Hertfordshire_LSIP_Main_Tables",
    "Hull_and_East_Yorkshire_LSIP_Main_Tables",
    "Kent_and_Medway_LSIP_Main_Tables",
    "Lancashire_LSIP_Main_Tables",
    "Leicester_and_Leicestershire_LSIP_Main_Tables",
    "Liverpool_City_Region_LSIP_Main_Tables",
    "Norfolk_and_Suffolk_LSIP_Main_Tables",
    "North_East_LSIP_Main_Tables",
    "North_of_Tyne_LSIP_Main_Tables",
    "Oxfordshire_LSIP_Main_Tables",
    "Solent_LSIP_Main_Tables",
    "South_East_Midlands_LSIP_Main_Tables",
    "South_Yorkshire_LSIP_Main_Tables",
    "Stoke-on-Trent_and_Staffordshire_LSIP_Main_Tables",
    "Swindon_and_Wiltshire_LSIP_Main_Tables",
    "Tees_Valley_LSIP_Main_Tables",
    "Thames_Valley_Berkshire_LSIP_Main_Tables",
    "The_Marches_LSIP_Main_Tables",
    "West_Midlands_and_Warwickshire_LSIP_Main_Tables",
    "West_of_England_and_North_Somerset_LSIP_Main_Tables",
    "West_Yorkshire_LSIP_Main_Tables",
    "Worcestershire_LSIP_Main_Tables",
    "York_and_North_Yorkshire_LSIP_Main_Tables"
]

# Initialize an empty list to store data for all files
all_processed_data = []

# Iterate over the file base names and process each file
for base_name in file_base_names:
    # Generate the full file path by adding the file extension
    file_path = os.path.join(f"{base_name}.ods")

    if os.path.exists(file_path):  # Check if the file exists
        try:
            # Load the spreadsheet
            xls = pd.read_excel(file_path, sheet_name="Occ_T5", engine="odf")

            # Extract the required table (Q4:AE16)
            selected_data = xls.iloc[3:16, 32:47]  # Convert Excel's column letters to zero-based indices

            # Rename columns based on the provided structure
            selected_data.columns = [
                "Category", "2015_FT", "2015_PT", "2015_SE", "Total_2015", "", "2020_FT", "2020_PT", "2020_SE", "Total_2020", "", "2035_FT", "2035_PT", "2035_SE", "Total_2035"
            ]

            # Drop empty columns
            selected_data = selected_data.drop(columns=["", "Total_2015", "Total_2020", "Total_2035"], errors='ignore')

            # Transpose the data
            transposed_data = selected_data.T

            # Reset the index to move index into a column
            transposed_data.reset_index(inplace=True)

            # Set the first row (index 0) as the header and drop it from the DataFrame
            transposed_data.columns = transposed_data.iloc[0]
            transposed_data = transposed_data.drop(index=0)

            # Drop columns
            transposed_data = transposed_data.drop(transposed_data.columns[[1, 2, 3, -1]], axis=1)

            # Remove 'Category' column and create 'Year' and 'Employment_Type' columns
            transposed_data['Year'] = transposed_data['Category'].str.extract(r'(\d{4})')  # Extract the year part
            transposed_data['Employment_Type'] = transposed_data['Category'].str.extract(r'_(\w+)')

            # Drop the 'Category' column as it's now split into 'Year' and 'Employment_Type'
            transposed_data.drop(columns=['Category'], inplace=True)

            # Extract region name from the file path
            region_name = os.path.splitext(os.path.basename(file_path))[0].replace("_LSIP_Main_Tables", "")

            # Add the 'Region' column at the first position with the extracted region name
            transposed_data.insert(0, "Region", region_name)

            # Add 'Gender' and 'Type' columns with the constant values
            transposed_data['Gender'] = 'Female'
            transposed_data['Type'] = 'Levels (000s)'

            # Reorder columns
            column_order = ['Region', 'Gender', 'Year', 'Type', 'Employment_Type'] + [col for col in transposed_data.columns if col not in ['Region', 'Gender', 'Year', 'Type', 'Employment_Type']]
            transposed_data = transposed_data[column_order]

            # Append the processed data to the list
            all_processed_data.append(transposed_data)

        except Exception as e:
            print(f"Error loading file {file_path}: {e}")
    else:
        print(f"File not found: {file_path}")

# Combine all the processed data into a single DataFrame
combined_data = pd.concat(all_processed_data, ignore_index=True)

# Display the combined data
combined_data
```

```python
# Save the DataFrame as an Excel file
combined_data.to_excel("Females_levels.xlsx", index=False)
```
